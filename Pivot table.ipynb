```python
import pandas as pd
```

```python
df=pd.read_excel('data.xlsx')
df.head()
```

```python
pivot_table = df.pivot_table(index='Country', columns='Product', values='Sales', 
aggfunc='sum')
pivot_table
```

```python
pivot_table.sort_values (by='A', ascending=False)
```

```python
pivot_table2 = df.pivot_table(
index=['Country','Category'], #multiple index columns
columns=['Product'], #multiple columns for products
values=['Sales'], #values to aggregate
aggfunc=['sum','mean'] #multiple aggregation functions
)
pivot_table2
```
