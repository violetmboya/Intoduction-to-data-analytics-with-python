```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('data.xlsx')
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'])   ##convert to datetime format

## ploting the data
plt.plot(df['Date'], df['Sales'], color='yellow')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(df['Product'], df['Sales'], color='pink')
plt.title('Sales per Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
sales_data=df.groupby('Category')['Sales'].sum()  ##aggregate data by car model
sales_data
```

```python
##create the pie chart
plt.figure(figsize=(8,8))
plt.pie(sales_data, labels=sales_data.index, autopct='%1.1f%%')

plt.title('Production of Sales by Category')
plt.show()
```

```python
plt.hist(df['Sales'], bins=5, color='pink', edgecolor='red')
plt.title('Max Price (Lakh) vs Insurance')
plt.xlabel('Sales')
plt.ylabel('Insurance')
plt.show()
```

```python
plt.fill_between(df['Date'], df['Sales'], color='lightgreen', alpha=0.1)
plt.plot(df['Date'], df['Sales'], color='green')
plt.title('Cumulative Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()                 
```
