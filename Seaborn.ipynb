```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car=pd.read_excel("C:/Users/HP/Downloads/Car data saved.xlsx")
car.head()
```

```python
variables= ['Min Price (Lakh)',	'Max Price (Lakh)',	'Range (kmpl)',	'CC', 
            'Seats', 'Variants', 'Ex-Showroom Price',	'RTO',	'Insurance', 
            'Other', 'Onroad Price']
```

```python
plt.figure(figsize=(15,12))
for i, var in enumerate(variables, start=1):
    plt.subplot(4, 3, i)
    sns.scatterplot(x=car[var], y=car['Max Price (Lakh)'], alpha=0.7)
    plt.title(f'Scatter plot of {var} vs Onroad Price')
    plt.xlabel(var)
    plt.ylabel('Target')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))
for i, var in enumerate(variables, start=1):
    plt.subplot(4, 3, i)
    sns.histplot(car[var], bins=20, kde=True, color='violet')
    plt.title(f'Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
